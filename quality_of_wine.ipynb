```python
X = dataframe.drop(columns = 'quality',axis =1)
Y = dataframe['quality']
```

```python
fig ,ax = plt.subplots(figsize=(15,15))
sns.boxplot(data = X,ax =ax)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
fig,ax = plt.subplots(figsize = (15,15))
sns.boxplot(data = X_scaled ,ax = ax)
```

```python
cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH','sulphates','alcohol','Id']
for col in cols:
    q1 = X[col].quantile(0.25)
    q3 = X[col].quantile(0.75)
    Iq = q3 - q1
    lower_bound  = q1 - 1.5*Iq
    Upper_bound =  q3 + 1.5*Iq
    mask = (X[col] >= lower_bound) & ( X[col] <= Upper_bound)
```

```python
X_outlier_detect = X[mask]
Y_outlier_detect  = Y[mask]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_outlier_detect)
```

```python
cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH','sulphates','alcohol','Id']
```

```python
X_scaled = pd.DataFrame(X_scaled,columns = cols)
X_scaled.describe()
```

```python
X_scaled.reset_index(drop=True, inplace=True)
Y_outlier_detect.reset_index(drop=True, inplace=True)
```

```python
q = X_scaled['chlorides'].quantile(.95)
mask = X_scaled['chlorides'] < q
dataNew = X_scaled[mask]
Y_outlier_detect = Y_outlier_detect[mask]
```

```python
fig,ax = plt.subplots(figsize = (15,15))
sns.boxplot(data = dataNew,ax =ax)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split( dataNew,Y_outlier_detect , test_size =  0.33,random_state = 52)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
accaurate_pred = accuracy_score(y_test,y_predict)
print(accaurate_pred)
```
